```python
%matplotlib notebook
import matplotlib.pyplot as plot
dm=app_train['AMT_INCOME_TOTAL'].mean()
ds=app_train['AMT_INCOME_TOTAL'].std()
fig = plt.figure('直方圖')
plt.hist(dm, bins=10, edgecolor="LightSteelBlue") 
```

```python
plt.subplots(4,1,figsize=(10,10))
AMT_INCOME_TOTAL=plt.subplot()
AMT_INCOME_TOTAL=app_train['AMT_INCOME_TOTAL'].hist(bins=30,color="LIMEGREEN",edgecolor="BLACK",alpha=0.7,histtype='bar',range = (8000,120000))
AMT_INCOME_TOTAL.set_xlabel("AMT_INCOME_TOTAL")
AMT_INCOME_TOTAL.set_ylabel("Frequency")
```
